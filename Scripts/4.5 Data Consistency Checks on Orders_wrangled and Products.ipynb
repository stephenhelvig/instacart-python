# Orders.csv - further data cleaning

```python
# Consistency check for df_orders
df_orders.describe()
```

Everything looks good here with the exception that 'Unnamed' may just be an index column. The count for days_since_prior_order is lower than the rest of the columns. This likely means that when there is no 2nd order yet, the entry stays blank. 

```python
# check if all the values in 'Unnamed:0' are the same as the index column
(df_orders.index == df_orders['Unnamed: 0']).all()
```

```python
# drop the 'Unnamed: 0' column
df_orders = df_orders.drop(columns=['Unnamed: 0'])
```

```python
df_orders.describe()
```

```python
# check for mixed data types
for col in df_orders.columns.tolist():
	weird = (df_orders[[col]].map(type) != df_orders[[col]].iloc[0].apply(type)).any(axis = 1)
	if len (df_orders[weird]) > 0:
		print(col)
```

```python
# check for missing values
df_orders.isnull().sum()
```

There are 206,209 missing values in 'days_since_prior_order'. It's possible this column doesn't perform a calculation until another order is placed and therefore, is still empty from customers that have only purchased from Instacart a single time.

```python
# create a new flag column (boolean) to address if the order is a first time order or not
df_orders['first_order'] = df_orders['days_since_prior_order'].isna()
```

```python
df_orders
```

I used a flag column to deal with these missing values because removing them or imputing them are not good options. 

```python
# Find duplicates
df_dups = df_orders[df_orders.duplicated()]
```
