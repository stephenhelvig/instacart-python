# Data Consistency checks: customers.csv

```python
# renaming columns in customer df
cust = cust.rename(columns={
    'First Name' : 'first_name',
    'Surnam' : 'last_name',
    'Gender' : 'gender',
    'STATE' : 'state',
    'Age' : 'age',
    'n_dependants' : 'number_of_dependents',
    'fam_status' : 'family_status'})
```
