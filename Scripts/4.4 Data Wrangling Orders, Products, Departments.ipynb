```python
# Find frequency of 'order_hour_of_day (count how many orders each hour)
df_orders['order_hour_of_day'].value_counts()
```

```python
# Count how many orders per hour
hourly_counts = df_orders['order_hour_of_day'].value_counts().sort_index()

# Find the busiest hour
busiest_hour = hourly_counts.idxmax()
busiest_count = hourly_counts.max()

print("Busiest hour:", busiest_hour)
print("Number of orders:", busiest_count)
```

# Subsetting Practice: Using .isin()

```python
# The sales team in your client’s organization wants to know more about breakfast item sales. Create a subset containing only the required information.
df_breakfast = df_products.loc[df_products['department_id'].isin([14])]
```

```python
df_breakfast
```

```python
# They’d also like to see details about products that customers might use to throw dinner parties. Your task is to find all observations from the entire dataframe that include items from the following departments: alcohol, deli, beverages, and meat/seafood. You’ll need to present this subset to your client.
df_dinnerparties = df_products.loc[df_products['department_id'].isin([5,7,12,20])]
```
